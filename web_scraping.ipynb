```python
import requests
from bs4 import BeautifulSoup

def scrape_category_page(url):
    """ Nos conectamos a la pagina,
    guardamos el titulo y el precio de todos los libros
    los presentamos.
    """
    response = requests.get(url)
    # lanza error si status != 200
    response.raise_for_status()
    response.encoding = "utf-8"
    # convierte el HTML en un objeto que podés recorrer con selectores.
    soup = BeautifulSoup(response.text, "html.parser")
    libros = []
    # product_pod representa cada libro
    for card in soup.select(".product_pod"):
        title = card.h3.a["title"]
        price = card.select_one(".price_color").text.strip()
        libros.append((title, price))
    return libros

# ejemplo: categoría Travel
url = "http://books.toscrape.com/catalogue/category/books/travel_2/index.html"
print(scrape_category_page(url))
```

```python
import requests
from bs4 import BeautifulSoup

BASE_URL = "http://books.toscrape.com/"

def get_categories():
    """Descarga la página principal y devuelve
    una lista de (nombre, url) de todas las categorías.
    """
    response = requests.get(BASE_URL)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, "html.parser")
    
    categorias = []
    # el menú lateral tiene las categorías
    for a in soup.select(".side_categories a"):
        name = a.text.strip()
        href = a["href"]
        # armamos la URL completa
        full_url = BASE_URL + href
        categorias.append((name, full_url))
    return categorias

print(get_categories())
```

```python
def scrape_all_categories():
    """Recorre todas las categorías y devuelve
    una lista con (categoria, titulo, precio).
    """
    categorias = get_categories()
    todos_libros = []
    for name, url in categorias:
        try:
            libros = scrape_category_page(url)
            for titulo, precio in libros:
                todos_libros.append((name, titulo, precio))
        except Exception as e:
            print(f"Error en categoría {name}: {e}")
    return todos_libros

libros = scrape_all_categories()
# mostramos solo los primeros 10
print(libros[:10])

```
